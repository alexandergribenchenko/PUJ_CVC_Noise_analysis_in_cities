```python
# Transformaciones requeridas (comunas para la transformación)
columnas_transformacion_01 = ['AreaM2', 'AREAHABITA', 'POBLACION','LDN_DOM', 'LDN_ORD', 'LAeq_DD', 'LAeq_DN', 'LAeq_OD', 'LAeq_ON']
columnas_transformacion_02 = ['AreaM2', 'AREAHABITA', 'POBLACION']
columnas_transformacion_03 = ['LDN_DOM', 'LDN_ORD', 'LAeq_DD', 'LAeq_DN', 'LAeq_OD', 'LAeq_ON']
```

```python
# Transformaciones requeridas (transformaciones implementadas)
df[columnas_transformacion_01] = df[columnas_transformacion_01].applymap(lambda x: float(x.replace(",",".")))
df[columnas_transformacion_02] = df[columnas_transformacion_02].applymap(lambda x: round(x,2))
df[columnas_transformacion_03] = df[columnas_transformacion_03].applymap(lambda x: round(x,1))
df
```

### Discriminación por intervalos de nivel de presión sonora

```python
# Generamos los intervalos y las categorias en los que queremos etiquetar las mediciones
bins = pd.IntervalIndex.from_tuples([(0, 35), (35, 40), (40, 45), (45, 50), (50, 55),(55, 60),(60, 65),(65, 70),(70, 75),(75, 80), (80, 140)], closed='left')
categorias = [' < 35','35 - 40', '40 - 45', '45 - 50', '50 - 55', '55 - 60', '60 - 65', '65 - 70', '70 - 75', '75 - 80','80 >']
print(bins)
print(categorias)
```

```python
# Se muestra a manera de ejemplo como se haría la etiqueta de una columna en especifico
pd.Series(np.array(categorias)[pd.cut(df["LDN_DOM"], bins = bins).cat.codes], index=df.index)
```

```python
# Se define la totalidad de las columnas a las que se les hará el etiquetado
columnas_etiquetado = ['LDN_DOM', 'LDN_ORD', 'LAeq_DD', 'LAeq_DN', 'LAeq_OD', 'LAeq_ON']
columnas_etiquetado
```

```python
# Se generara una columna adicional en el df para cada columna que requiera etiquetado
for indicador in columnas_etiquetado: 
    df[indicador+"_cat"] = pd.Series(np.array(categorias)[pd.cut(df[indicador], bins = bins).cat.codes], index=df.index)
```

```python
df
```

### Tablas de resultados: Municipio

```python
def municipio_porcentaje(df, indicador, ICAU): 
    salida = df.groupby(indicador+'_cat').sum()['POBLACION'].to_frame().reset_index().rename(columns={"POBLACION": "Población expuesta (personas)",indicador+'_cat': indicador})
    salida['Población expuesta (%)'] = round(100*salida['Población expuesta (personas)']/salida['Población expuesta (personas)'].sum(),1)
    salida = salida.drop('Población expuesta (personas)',axis=1)
    salida = salida.transpose()
    salida.columns = list(salida.loc[indicador,:])
    salida = salida.drop(indicador,axis=0)
    if ICAU == True: 
            salida['PUAR']= salida.loc[:,'65 - 70':'80 >'].sum(axis=1)
    salida=salida.applymap(lambda x: round(x,3))
    return salida
```

#### Municipio: Ordinario Porcentaje

```python
municipio_porcentaje(df,'LDN_ORD', ICAU=True).transpose().reset_index().rename(columns={'index': 'NPS (dBA)'})
```

```python
# Funcion que nos permite definir la relación de color con 
def colors_from_values(values, palette_name):
    # normalize the values to range [0, 1]
    normalized = (values - min(values)) / (max(values) - min(values))
    # convert to indices
    indices = np.round(normalized * (len(values) - 1)).astype(np.int32)
    # use the indices to get the colors
    palette = sns.color_palette(palette_name, len(values))
    return np.array(palette).take(indices, axis=0)
```

```python
# Visualización gráfica
plt.figure(figsize=(8,5))
sns.barplot(x='Población expuesta (%)',
            y='index',
            data=municipio_porcentaje(df,'LDN_ORD', ICAU=True).transpose().reset_index(), 
            #palette=colors_from_values(municipio_porcentaje(df,'LDN_ORD', ICAU=True).transpose().reset_index()['Población expuesta (%)'], "coolwarm"))
            palette="coolwarm")
plt.title('Distribución de población expuesta a ruido - Buga - Jornada Ordinaria')
plt.ylabel('Nivel de presión sonora [dBA]')
plt.xlabel('Porcentaje de población expuesta [%]')
plt.show()
```

#### Municipio: Ordinario cantidad

```python
def municipio_cantidad(df, indicador, ICAU): 
    salida = df.groupby(indicador+'_cat').sum()['POBLACION'].to_frame().reset_index().rename(columns={"POBLACION": "Población expuesta (personas)",indicador+'_cat': indicador})
    salida['Población expuesta (personas)'] = salida['Población expuesta (personas)'].apply(lambda x: int(x))
    salida = salida.transpose()
    salida.columns = list(salida.loc[indicador,:])
    salida = salida.drop(indicador,axis=0)
    if ICAU == True: 
            salida['Poblacion expuesta (>65 dBA)']= salida.loc[:,'65 - 70':'80 >'].sum(axis=1).apply(lambda x: int(x))
    return salida
```

```python
municipio_cantidad(df,'LDN_ORD', ICAU=True).transpose().reset_index().rename(columns={'index': 'NPS (dBA)'})
```

```python
# Visualización gráfica
plt.figure(figsize=(8,5))
sns.barplot(x='Población expuesta (personas)',
            y='index',
            data=municipio_cantidad(df,'LDN_ORD', ICAU=True).transpose().reset_index(), 
            #palette=colors_from_values(municipio_porcentaje(df,'LDN_ORD', ICAU=True).transpose().reset_index()['Población expuesta (%)'], "coolwarm"))
            palette="coolwarm")
plt.title('Distribución de población expuesta a ruido - Buga - Jornada Ordinaria')
plt.ylabel('Nivel de presión sonora [dBA]')
plt.xlabel('Cantidad de personas expuestas')
plt.show()
```

### Resultados: Comunas

#### Comunas porcentaje

```python
def comunas_porcentaje(df, indicador, ICAU):
    salida_total = pd.DataFrame() 
    comunas = sorted(df.COMUNA.unique())
    for com in comunas:
        df_com = df[df.COMUNA==com]
        salida_com = municipio_porcentaje(df_com,indicador,ICAU)
        salida_total = pd.concat([salida_total,salida_com])
    salida_total = salida_total.fillna(0)
    salida_total.index=comunas
    salida_total=salida_total.reindex(sorted(salida_total.columns), axis=1)
    return salida_total
```

```python
comunas_porcentaje(df,'LDN_ORD', ICAU=True)
```

```python
# Visualización gráfica
plt.figure(figsize=(8,5))
sns.barplot(x='PUAR',y='index',
            data=comunas_porcentaje(df,'LDN_ORD', ICAU=True).reset_index(), 
            palette=colors_from_values(comunas_porcentaje(df,'LDN_ORD', ICAU=True).reset_index()['PUAR'], "coolwarm"))
plt.title('Distribución de población expuesta por comuna - Buga - Jornada Ordinaria')
plt.ylabel('Comuna')
plt.xlabel('PUAR [%]')
plt.show()
```
